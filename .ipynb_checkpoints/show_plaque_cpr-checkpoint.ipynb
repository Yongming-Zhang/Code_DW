```python
import matplotlib.pyplot as plt
#%matplotlib inline
from PIL import Image
import numpy as np
import cv2 
import torch
import re 
import os

name = '1036604/A4CC76BE/65669469_CTA'
origin_cpr_path = '/mnt/DrwiseDataNFS/drwise_runtime_env/data1/inputdata/'+name+'/diagnose_debug/segmentation'
name = '1036604_60_0416/A4CC76BE/65669469_CTA'
new_cpr_path = '/mnt/DrwiseDataNFS/drwise_runtime_env/data1/inputdata/'+name+'/diagnose_debug/segmentation'

_nsre = re.compile('([0-9]+)')
def natural_sort_key(s):
    return [int(text) if text.isdigit() else text.lower() for text in re.split(_nsre, s)]

def plot_imgs(item_dir, num_imgs, step):
    all_item_dirs = sorted(os.listdir(item_dir), key=natural_sort_key)
    item_files = [os.path.join(item_dir, file) for file in all_item_dirs][num_imgs:num_imgs+step]

    plt.figure(figsize=(100, 100))
    for idx, img_path in enumerate(item_files):
        plt.subplot(4, 5, idx+1)
        print(img_path)
        img = plt.imread(img_path)
        plt.imshow(img)#, cmap='gray'
    plt.tight_layout()
plot_imgs(origin_cpr_path, num_imgs=100, step=10)

def plot_imgs1(item_dir, num_imgs, step):
    all_item_dirs = sorted(os.listdir(item_dir), key=natural_sort_key)
    item_files = [os.path.join(item_dir, file) for file in all_item_dirs][num_imgs:num_imgs+step:3]

    plt.figure(figsize=(100, 100))
    for idx, img_path in enumerate(item_files):
        plt.subplot(4, 5, idx+1)
        print(img_path)
        img = plt.imread(img_path)
        plt.imshow(img)#, cmap='gray'
    plt.tight_layout()
plot_imgs1(new_cpr_path, num_imgs=300, step=30)
```
